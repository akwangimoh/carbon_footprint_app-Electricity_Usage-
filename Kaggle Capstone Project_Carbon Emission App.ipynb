```python
intake_agent = Agent(
    name="electricity_intake_agent",
    description="Asks the user for electricity usage in kWh and month, calculates emissions.",
    model=Gemini(model="gemini-2.5-flash-lite"),
    instruction="""
Welcome the user to the TCN Carbon Footprint App for Electricity Usage!
Ask the user for their electricity usage in kWh and the month.
After receiving both, calculate CO2 emissions using 0.402 kg CO2 per kWh.
Reply with: 'For X kWh in [month], the estimated CO2 emissions are Y kg.'
Only respond after both pieces of information are received.
"""
)
print("Agent Created")
```

```python
from google.adk.runners import InMemoryRunner
```

```python
runner = InMemoryRunner(agent=intake_agent)
print("✅ Runner created successfully.")
```

```python
response = await runner.run_debug(
"Please Welcome the user to TCN Carbon Foot Print App for Electricity Usage, and ask the user to type their electricity usage in kWh and the month."
)
```

```python
# Inspect the first event to see its attributes
print(response[0])
```

```python
# Loop through all events and print their text
for event in response:
    if hasattr(event, "content") and event.content.parts:
        print(event.content.parts[0].text)
```

```python
runner = InMemoryRunner(agent=intake_agent)
print("✅ Runner created")
```

```python
# Step 1: Display welcome message via agent
welcome_response = await runner.run_debug(
    "Welcome the user to the TCN Carbon Footprint App for Electricity Usage! "
    "Ask the user to type their electricity usage in kWh and the month."
)

# Print the agent's welcome message
for event in welcome_response:
    if hasattr(event, "content") and event.content.parts:
        print(event.content.parts[0].text)

# Step 2: Ask user for input
kwh = input("Please enter your electricity usage in kWh: ")
month = input("Please enter the month: ")

# Step 3: Combine the inputs into a single message for the agent
user_input = f"{kwh} kWh, {month}"

# Step 4: Send user input to the agent
response = await runner.run_debug(user_input)

# Step 5: Print the agent's output
for event in response:
    if hasattr(event, "content") and event.content.parts:
        print(event.content.parts[0].text)
```
